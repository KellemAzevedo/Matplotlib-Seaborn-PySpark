# Haetmap com correlação Seaborn

```python
'''
Sempre que desejamos avaliar a relação de uma variável quantitativa com outra
quantitativa, podemos calcular o coeficiente de correlação que é um número que
varia de -1 até 1 onde:
* 1 significa perfeita correlação positiva, ou seja, quando x aumenta, y aumenta na mesma proporção;
* -1 significa perfeita correlação negativa, ou seja, quando x aumenta, y diminui na mesma proporção;
* aprox 1 significa forte correlação positiva, ou seja, quando x aumenta, y tende a aumentar na mesma proporção;
* aprox -1 significa forte correlação negativa, ou seja, quando x aumenta, y tende a diminuir na mesma proporção;
* aprox 0 correlação não significativa
Nesses casos além da medida do coeficiente de correlação em si, podemos avaliar a relação quantitativa entre duas variáveis
por gráficos, sendo o diagrama de dispersão, o mais comum deles, mas também podemos utilizar o heatmap de correlação,
este por sua vez, permite colocar vários pares de atributos quantitativos em uma matriz onde cada célula corresponde
a um par de variáveis (x,y)
'''
```

```python
sns.heatmap(df_log[['Distância', 'Tempo de Entrega']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlação entre Distância e Tempo de entrega')
plt.show()
```

```python
propaganda = [1,2,3,4,5]
vendas = [8,5,4,2,1]
df = pd.DataFrame(zip(propaganda,vendas),columns=['propaganda','vendas'])
sns.heatmap(df[['propaganda', 'vendas']].corr(), annot=True)
plt.title('Correlação entre Propaganda e Vendas')
plt.show()
```

```python
propaganda = [1,2,3,4,5]
vendas = [1,2,4,5,8]
investimento = [10,19,29,43,57]
df = pd.DataFrame(zip(propaganda,vendas,investimento),columns=['propaganda','vendas','investimento'])
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlação entre Propaganda e Vendas')
plt.show()
```

```python
# Converter coluna de data para tipo datetime
df_fin['Data'] = pd.to_datetime(df_fin['Data'])

# Filtrar por tipo de transação
receita = df_fin[df_fin['Tipo de Transação'] == 'Receita']
despesa = df_fin[df_fin['Tipo de Transação'] == 'Despesa']

# Agrupar por mês e calcular a soma de receitas e despesas
monthly_revenue = receita.groupby(df_fin['Data'].dt.month)['Valor'].sum()
monthly_expense = despesa.groupby(df_fin['Data'].dt.month)['Valor'].sum()

# Plotar gráfico de barras agrupadas de receitas e despesas por mês
monthly_revenue.plot(kind='bar', color='lightgreen', position=0, width=0.4, label='Receitas')
monthly_expense.plot(kind='bar', color='lightblue', position=1, width=0.4, label='Despesas')
plt.xlabel('Mês')
plt.ylabel('Valor')
plt.title('Comparação de Receitas e Despesas por Mês')
plt.legend()
plt.xticks(rotation=0)
plt.show()
```

```python
# Converter coluna de data para tipo datetime
df_fin['Data'] = pd.to_datetime(df_fin['Data'])

# Calcular saldo mensal (receitas - despesas)
monthly_balance = df_fin.groupby(df_fin['Data'].dt.month)['Valor'].sum()

# Plotar gráfico de linhas da tendência do saldo mensal
monthly_balance.plot(marker='o', linestyle='-', color='purple')
plt.xlabel('Mês')
plt.ylabel('Saldo')
plt.title('Tendência do Saldo Mensal')
plt.grid(True)
plt.show()
```

```python
# Converter coluna de data para tipo datetime
df_fin['Data'] = pd.to_datetime(df_fin['Data'])

# Filtrar por tipo de transação
revenue_data = df_fin[df_fin['Tipo de Transação'] == 'Receita']

# Calcular receitas acumuladas por categoria ao longo do tempo
cumulative_revenue = revenue_data.groupby([revenue_data['Data'].dt.month, 'Categoria'])['Valor'].sum().unstack().cumsum()

# Plotar gráfico de área da distribuição acumulada de receitas por categoria ao longo do tempo
cumulative_revenue.plot(kind='area', cmap='coolwarm')
plt.xlabel('Mês')
plt.ylabel('Receita Acumulada')
plt.title('Distribuição Acumulada de Receitas por Categoria ao Longo do Tempo')
plt.legend(title='Categoria')
plt.grid(True)
plt.show()
```

```python
import numpy as np
'''
Acertar a comparação com bases de mesmo tamanho
'''
# Calcular média de valor por categoria
avg_data = df_fin.groupby('Categoria').agg({'Valor':'mean'}).reset_index()

# Normalizar os dados
avg_data['Valor'] = avg_data['Valor'] / avg_data['Valor'].max()

# Plotar gráfico de radar
labels = avg_data['Categoria']
stats = avg_data['Valor'].tolist()
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

stats = np.concatenate((stats,[stats[0]]))
angles = np.concatenate((angles,[angles[0]]))

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='green', alpha=0.25)
ax.plot(angles, stats, color='green', linewidth=2)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=12)

plt.title('Comparação de Média de Valor por Categoria', size=20, color='green', y=1.1)
plt.show()
```
